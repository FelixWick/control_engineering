```python
import numpy as np
import matplotlib.pyplot as plt
import control as ct
```

```python
A = [[0, 1], [-1, -0.2]]
B = [[0], [1]]
C = [[1, 0]]
D = [[0]]
```

```python
K = ct.place(A, B, [-2, -3])
```

```python
sys_cl = ct.ss(A - np.dot(B, K), B, C, D)
```

```python
t, y = ct.step_response(sys_cl)
```

```python
plt.grid()
plt.plot(t, y)
```
