```python
import numpy as np
import matplotlib.pyplot as plt
import control as ct
```

```python
A = [[0, 1], [-1, -0.2]]
B = [[0], [1]]
C = [[1, 0]]
D = [[0]]

sys_ss = ct.ss(A, B, C, D)
sys_tf = ct.tf(sys_ss)
```

```python
Q = np.eye(2)
R = [[1]]
```

```python
K, _, _ = ct.lqr(sys_ss, Q, R)
```

```python
sys_lqr = ct.ss(A - B @ K, B, C, D)
```

```python
t, y = ct.step_response(sys_lqr)
```

```python
plt.grid()
plt.plot(t, y)
```
