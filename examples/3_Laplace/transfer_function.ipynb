```python
import numpy as np
import matplotlib.pyplot as plt
import control as ct
```

```python
G1 = ct.tf([1], [1, 1])
G2 = ct.tf([1], [1, 0.2, 1])
```

```python
t = np.linspace(0, 10, 500)
```

```python
_, y1 = ct.step_response(G1, t)
_, y2 = ct.step_response(G2, t)
```

```python
plt.figure()
plt.plot(t, y1, label="First-order")
plt.plot(t, y2, label="Second-order")
plt.legend()
plt.grid()
plt.show()
```
