```python
import matplotlib.pyplot as plt
import control as ct
```

```python
G = ct.tf([1], [1, 1])
Kps = [0.5, 1, 5]
```

```python
plt.figure()
for Kp in Kps:
    C = ct.tf([Kp], [1])
    T = ct.feedback(C * G)
    t, y = ct.step_response(T)
    plt.plot(t, y, label=f"Kp = {Kp}")

plt.xlabel("Time [s]")
plt.ylabel("Output")
plt.legend()
plt.grid()
plt.show()
```
