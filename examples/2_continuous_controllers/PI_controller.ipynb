```python
import numpy as np
import matplotlib.pyplot as plt
import control as ct
```

```python
G = ct.tf([1], [1, 1])
```

```python
C_P = ct.tf([2], [1])
C_PI = ct.tf([2, 1], [1, 0])
```

```python
T_P = ct.feedback(C_P * G)
T_PI = ct.feedback(C_PI * G)
```

```python
# Common time vector
t = np.linspace(0, 10, 500)

# reason:
# step_response() auto-generates time vectors
# different systems â†’ different time grids
```

```python
_, y_p = ct.step_response(T_P, t)
_, y_pi = ct.step_response(T_PI, t)
```

```python
plt.figure()
plt.plot(t, y_p, label="P control")
plt.plot(t, y_pi, label="PI control")
plt.xlabel("Time [s]")
plt.ylabel("Output")
plt.legend()
plt.grid()
plt.show()
```

PI controller eliminates steady-state error
