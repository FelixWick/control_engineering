```python
import matplotlib.pyplot as plt
import control as ct
```

```python
G = ct.tf([1], [1, 1])
Ts = [0.1, 0.5, 1.0]
```

```python
plt.figure()
for T in Ts:
    Gd = ct.sample_system(G, T, method="zoh")
    t, y = ct.step_response(Gd)
    plt.step(t, y, where="post", label=f"Ts = {T}")

plt.legend()
plt.grid()
plt.show()
```
