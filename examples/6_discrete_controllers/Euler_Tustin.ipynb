```python
import matplotlib.pyplot as plt
import control as ct
```

```python
G = ct.tf([1], [1, 1])
T = 0.5
```

```python
Gd_euler = ct.sample_system(G, T, method="euler")
Gd_tustin = ct.sample_system(G, T, method="tustin")
```

```python
t, y_e = ct.step_response(Gd_euler)
t, y_t = ct.step_response(Gd_tustin)
```

```python
plt.step(t, y_e, where="post", label="Euler")
plt.step(t, y_t, where="post", label="Tustin")
plt.legend()
plt.grid()
plt.show()
```
