```python
import matplotlib.pyplot as plt
import control as ct
```

```python
zetas = [0.2, 0.5, 1.0, 2.0, 5.0]

plt.figure()
for zeta in zetas:
    G = ct.tf([1], [1, 2*zeta, 1])
    t, y = ct.step_response(G)
    plt.plot(t, y, label=f" = {zeta}")

plt.xlabel("Time [s]")
plt.ylabel("Output")
plt.legend()
plt.grid()
plt.xlim(0, 30)
plt.show()
```
