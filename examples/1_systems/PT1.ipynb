```python
import matplotlib.pyplot as plt
import control as ct
```

```python
taus = [0.5, 1.0, 2.0]

plt.figure()
for tau in taus:
    G = ct.tf([1], [tau, 1])
    t, y = ct.step_response(G)
    plt.plot(t, y, label=f"Ï„ = {tau}")

plt.xlabel("Time [s]")
plt.ylabel("Output")
plt.legend()
plt.grid()
plt.show()
```
