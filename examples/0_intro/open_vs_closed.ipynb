```python
import numpy as np
import matplotlib.pyplot as plt
import control as ct
```

```python
# Plant: first-order system
G = ct.tf([1], [1, 1])
```

```python
# Open-loop response
t = np.linspace(0, 10, 500)
t_ol, y_ol = ct.step_response(G, t)
```

```python
# Closed-loop with proportional controller
Kp = 2
C = ct.tf([Kp], [1])
T = ct.feedback(C * G)

t_cl, y_cl = ct.step_response(T, t)
```

```python
plt.figure()
plt.plot(t_ol, y_ol, label="Open-loop")
plt.plot(t_cl, y_cl, label="Closed-loop")
plt.xlabel("Time [s]")
plt.ylabel("Output")
plt.legend()
plt.grid()
plt.show()
```
