```python
import matplotlib.pyplot as plt
```

```python
# Open-loop prediction
A = 1
B = 1
x = 0
u = 1
```

```python
xs = []
for _ in range(10):
    x = A * x + B * u
    xs.append(x)
```

```python
plt.plot(xs, marker='o')
plt.xlabel("Prediction step")
plt.ylabel("State")
plt.grid()
plt.show()
```
