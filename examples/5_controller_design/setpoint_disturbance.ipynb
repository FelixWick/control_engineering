```python
import numpy as np
import matplotlib.pyplot as plt
import control as ct
```

```python
G = ct.tf([1], [1, 1])
C = ct.tf([2, 1], [1, 0])

T_r = ct.feedback(C * G)
T_d = ct.feedback(G, C)

t = np.linspace(0, 10, 500)
t, y_r = ct.step_response(T_r, t)
t, y_d = ct.step_response(T_d, t)

plt.figure()
plt.plot(t, y_r, label="Setpoint response")
plt.plot(t, y_d, label="Disturbance response")
plt.legend()
plt.grid()
plt.show()
```
