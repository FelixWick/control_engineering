```python
import numpy as np
import matplotlib.pyplot as plt
import control as ct
```

Define plant transfer function

```python
s = ct.TransferFunction.s
G = 1 / ((s + 1) * (s + 2) * (s + 3))
```

Example PID controller

```python
Kp = 10
Ki = 0
Kd = 0

C = ct.TransferFunction([Kd, Kp, Ki], [1, 0])
```

Closed-loop system

```python
T = ct.feedback(C * G)
```

Simulation time

```python
t = np.linspace(0, 20, 1000)
```

Step response

```python
t, y = ct.step_response(T, t)
```

Reference input

```python
r = np.ones_like(t)
```

Control error

```python
e = r - y
```

Time step

```python
dt = t[1] - t[0]
```

```python
plt.plot(t, y, label="Output")
plt.plot(t, r, '--', label="Reference")
plt.title("System Response")
plt.xlabel("Time")
plt.ylabel("Output")
plt.legend()
plt.grid(True)
plt.show()
```

T-Sum calculation

```python
T_sum = np.sum(np.abs(e)) * dt
print("T-Sum value:", T_sum)
```

```python
Kp = 36
Ki = 38
Kd = 8.5
C = ct.TransferFunction([Kd, Kp, Ki], [1, 0])
T = ct.feedback(C * G)
t, y = ct.step_response(T, t)
e = r - y
dt = t[1] - t[0]
plt.plot(t, y, label="Output")
plt.plot(t, r, '--', label="Reference")
plt.title("System Response")
plt.xlabel("Time")
plt.ylabel("Output")
plt.legend()
plt.grid(True)
plt.show()
T_sum = np.sum(np.abs(e)) * dt
print("T-Sum value:", T_sum)
```
