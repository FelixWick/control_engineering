```python
import numpy as np
import matplotlib.pyplot as plt
import control as ct
```

Define plant transfer function

```python
s = ct.TransferFunction.s
G = 1 / ((s + 1) * (s + 2) * (s + 3))
```

Find Ziegler-Nichols parameters experimentally by gradually increasing Kp of appled controller (setting Ki and Kd to zero) until oscillation

```python
Kp = 1
Ki = 0
Kd = 0
```

PID controller

```python
C = ct.TransferFunction([Kd, Kp, Ki], [1, 0])
```

Closed-loop system

```python
T = ct.feedback(C * G)
```

```python
t, y = ct.step_response(T)
```

```python
plt.plot(t, y)
plt.title("Step Response during Ziegler-Nichols Tuning")
plt.xlabel("Time")
plt.ylabel("Output")
plt.grid(True)
plt.show()
```

For comparison: Ku approximately equals gain margin, and Tu can then be calculated from crossover frequency.

```python
gm, pm, wg, wp = ct.margin(G)

Ku = gm
Tu = 2 * np.pi / wg
```

```python
Ku
```

```python
Tu
```

Ziegler-Nichols PID tuning

```python
Kp = 0.6 * Ku
Ti = Tu / 2
Td = Tu / 8
```

```python
Ki = Kp / Ti
Kd = Kp * Td
```

```python
print(Kp, Ki, Kd)
```

```python
C = ct.TransferFunction([Kd, Kp, Ki], [1, 0])
T = ct.feedback(C * G)
t, y = ct.step_response(T)
plt.plot(t, y)
plt.title("Step Response with Ziegler-Nichols PID")
plt.xlabel("Time")
plt.ylabel("Output")
plt.grid(True)
plt.show()
```
