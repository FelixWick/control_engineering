```python
import numpy as np
import matplotlib.pyplot as plt
import control as ct
```

```python
G = ct.tf([1], [1, 6, 11, 6])   # 1 / ((s+1)(s+2)(s+3))
```

```python
gm, pm, wg, wp = ct.margin(G)
```

```python
Ku = gm
Tu = 2 * np.pi / wg
```

```python
Kp = 0.6 * Ku
Ti = Tu / 2
Td = Tu / 8
```

```python
C = Kp * ct.tf([Ti*Td, Ti, 1], [Ti, 0])
T = ct.feedback(C * G)
```

```python
t, y = ct.step_response(T)
```

```python
plt.plot(t, y)
plt.grid()
plt.show()
```
